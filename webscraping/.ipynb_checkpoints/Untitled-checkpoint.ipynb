```python


def button_clicked(button):
    for other_button in buttons:
        other_button.config(state="normal", bg="SystemButtonFace")
    button.config(state="disabled", bg="gray")
    text.config(state="normal")
    text.delete(1.0, tk.END)
    text.insert(tk.END, button["text"])
    text.tag_add("center", "1.0", "end")
    text.config(state="disabled")
    run_button.config(state="normal")
    cancel_button.config(state="disabled")
    global cancel_id
    cancel_id = None

def run():
    label = text.get("1.0", tk.END).strip()
    output.insert(tk.END, f"{label} clicked.\n")
    driver = get_webpage(url)
    

def run_button_clicked():
    label = text.get("1.0", tk.END).strip()
    if label:
        output.config(state="normal")
        output.insert(tk.END, f"{label} clicked.\n")
        output.config(state="disabled")
        text.config(state="normal")
        text.config(state="disabled")
        run_button.config(state="disabled")
        cancel_button.config(state="normal")
        global cancel_id
        cancel_id = root.after(0, run)

def cancel_button_clicked():
    global cancel_id
    if cancel_id is not None:
        root.after_cancel(cancel_id)
    run_button.config(state="normal")
    cancel_button.config(state="disabled")
    output.config(state="normal")
    output.delete("1.0", tk.END)
    output.config(state="disabled")
    for button in buttons:
        button.config(state="normal", bg="SystemButtonFace")

MIN_WIDTH, MIN_HEIGHT = 510, 400

root = tk.Tk()
root.title("Shopee webscraper")
root.geometry(f"{MIN_WIDTH}x{MIN_HEIGHT}")
root.minsize(MIN_WIDTH, MIN_HEIGHT)

screen_width = root.winfo_screenwidth()
screen_height = root.winfo_screenheight()
x = (screen_width - MIN_WIDTH) // 2
y = (screen_height - MIN_HEIGHT) // 2
root.geometry(f"+{x}+{y}")

button_frame = tk.Frame(root)
button_frame.pack(fill="both", expand=True, padx=10, pady=10)

text = tk.Text(root, height=1, width=20, state="disabled")
text.pack(fill="x", padx=10, pady=(10, 5))
text.tag_configure("center", justify="center")

output = tk.Text(root, height=5, width=50, state="disabled")
output.pack(fill="both", expand=True, padx=10, pady=(5, 0))

scrollbar = tk.Scrollbar(output)
scrollbar.pack(side="right", fill="y")
output.config(yscrollcommand=scrollbar.set)
scrollbar.config(command=output.yview)

buttons = []

for i in range(len(button_labels)):
    button = tk.Button(button_frame, text=button_labels[i], width=20, height=2)
    button.grid(row=i // 3, column=i % 3, padx=5, pady=5)
    button.config(command=lambda button=button: button_clicked(button))
    buttons.append(button)

run_button = tk.Button(root, text="Run", width=10, height=2, state="disabled")
run_button.pack(side="left", padx=10, pady=10)

cancel_button = tk.Button(root, text="Reset", width=10, height=2, state="disabled")
cancel_button.pack(side="left", padx=5, pady=10)

exit_button = tk.Button(root, text="Exit", width=10, height=2, command=root.destroy)
exit_button.pack(side="right", padx=10, pady=10)

run_button.config(command=run_button_clicked)

cancel_button.config(command=cancel_button_clicked)

root.mainloop()
```
