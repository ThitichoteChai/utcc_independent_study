```python
import pandas as pd
import numpy as np
import os

file_name = 'data_raw.xlsx'
new_file_name = 'data.xlsx'

current_dir = os.getcwd()
data_path = os.path.join(current_dir, file_name)
data_save_path = os.path.join(current_dir, new_file_name)

print('file path', data_path)

print('new file path', data_save_path)

raw_data = pd.read_excel(data_path)
data = raw_data[['cat', 'comment']]

data.loc[:, 'creditability'] = ''
data.loc[:, 'content'] = ''
data.loc[:, 'expression'] = ''
```
