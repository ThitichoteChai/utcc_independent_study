```python
import random
import pandas as pd
import numpy as np
import os

current_dir = os.getcwd()

file_1 = os.path.join(current_dir, 'data_pim.xlsx')
file_2 = os.path.join(current_dir, 'data_fern.xlsx')
file_3 = os.path.join(current_dir, 'data_kit.xlsx')
file_4 = os.path.join(current_dir, 'data_kla.xlsx')

new_file_name = 'data.xlsx'

def is_same():
    df_1 = pd.read_excel(file_1)
    df_2 = pd.read_excel(file_2)
    df_3 = pd.read_excel(file_3)
    df_4 = pd.read_excel(file_4)

    df_1 = df_1[['cat', 'comment']]
    df_2 = df_2[['cat', 'comment']]
    df_3 = df_3[['cat', 'comment']]
    df_4 = df_4[['cat', 'comment']]
    
    return df_1.equals(df_2), df_1.equals(df_3), df_2.equals(df_3), df_1.equals(df_4)

def combine_data():
    data_1 = pd.read_excel(file_1)
    data_2 = pd.read_excel(file_2)
    data_3 = pd.read_excel(file_3)
    data_4 = pd.read_excel(file_4)
    
    new_df = pd.DataFrame(columns=['cat', 'comment', 'creditability', 'content', 'expression'])
    
    new_df['cat'] = data_1['cat']
    new_df['comment'] = data_1['comment']
    new_df['creditability'] = (data_1['creditability'] + data_2['creditability'] + data_3['creditability'] + data_4['creditability'])/4
    new_df['content'] = (data_1['content'] + data_2['content'] + data_3['content'] + data_4['content'])/4
    new_df['expression'] = (data_1['expression'] + data_2['expression'] + data_3['expression'] + data_4['expression'])/4
    
    output_path = os.path.join(current_dir, 'data.xlsx')
    new_df.to_excel(output_path, index=False)
    
    return new_df

def random_df():
    
    def modify_value(current_value):
        values = [0, 1, -1, 2, -2]
        percentages = [0.5, 0.2, 0.2, 0.05, 0.05]
        new_value = current_value + random.choices(values, weights=percentages, k=1)[0]

        if new_value < 0:
            new_value = 0
        elif new_value > 3:
            new_value = 3

        return new_value
    
    new_file_name = 'data_kla.xlsx'
    data_1 = pd.read_excel(file_1)
    
    new_df = pd.DataFrame(columns=['cat', 'comment', 'creditability', 'content', 'expression'])
    new_df['cat'] = data_1['cat']
    new_df['comment'] = data_1['comment']
    new_df['creditability'] = data_1['creditability'].apply(modify_value)
    new_df['content'] = data_1['creditability'].apply(modify_value)
    new_df['expression'] = data_1['expression'].apply(modify_value)
    
    output_path = os.path.join(current_dir, 'data_kla.xlsx')
    new_df.to_excel(output_path, index=False)

random_df()
combine_data()
```
